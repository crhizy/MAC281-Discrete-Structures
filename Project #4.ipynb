```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
C = nx.Graph()

C.add_edges_from([("President", "Vicepresident"),
                   ("Vicepresident", "OperatorManager"),
                   ("Vicepresident", "AdministrativeManager"),
                   ("AdministrativeManager", "Secretary"),
                   ("AdministrativeManager", "Counter"),
                   ("AdministrativeManager","PayrollOffice"),
                   ("AdministrativeManager","Clients"),
                   ("OperatorManager","Secretary"),
                   #("OperatorManager", "AdministrativeManager"),
                   ("OperatorManager", "HeadPressman"),
                   ("OperatorManager", "Suppliers"),
                   ("HeadPressman","PrePress")])
                
```

```python

nx.draw(C, with_labels= True, font_weight="bold")
plt.show()
```
